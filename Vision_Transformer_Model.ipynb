```python
import os
import torch
import numpy as np
from torch.utils.data import DataLoader, Dataset
import torchvision
import torchvision.transforms as T
from PIL import Image
import scipy.io
import matplotlib.pyplot as plt
import cv2
from torchvision import models

# Path to your dataset
images_folder = "/kaggle/input/stanford-cars-dataset/cars_train/cars_train/"

# Load annotations
annotations_path = "/kaggle/input/cars-devkit/cars_train_annos.mat"
cars_train_annos = scipy.io.loadmat(annotations_path)

# Extract 'annotations' array
annotations_list = cars_train_annos['annotations'][0]

# Load metadata file (contains class names)
meta_path = "/kaggle/input/cars-devkit/cars_meta.mat"
cars_meta = scipy.io.loadmat(meta_path)

# Extract class names properly
class_names = [name[0] for name in cars_meta['class_names'][0]]

# Define the transformation for image preprocessing
transform = T.Compose([
    T.Resize((224, 224)),  # Resize to the input size for ViT
    T.ToTensor(),
    T.Normalize(
        mean=[0.485, 0.456, 0.406],  # ImageNet mean values
        std=[0.229, 0.224, 0.225]    # ImageNet std values
    )
])

# Dataset class to load images and annotations
class CarsDataset(Dataset):
    def __init__(self, annotations, transform=None):
        self.annotations = annotations
        self.transform = transform

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, idx):
        anno = self.annotations[idx]
        
        # Correcting the way we handle the file name (anno['fname'] is an ndarray)
        fname = anno['fname'][0]  # Extract the filename properly as a string
        img_path = os.path.join(images_folder, fname)  # Join it with the image folder path
        image = Image.open(img_path)

        # Ensure the image is in RGB format
        if image.mode != 'RGB':
            image = image.convert('RGB')

        if self.transform:
            image = self.transform(image)

        # Adjusted to 'class' instead of 'class_id'
        label = anno['class'][0][0]  # Using 'class' to get the label
        return image, label

# Prepare DataLoader for the dataset
dataset = CarsDataset(annotations_list, transform=transform)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
import os
import torch
import numpy as np
from torch.utils.data import DataLoader, Dataset
import torchvision
import torchvision.transforms as T
from PIL import Image
import scipy.io
import matplotlib.pyplot as plt
import cv2
from torchvision import models
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import precision_score, recall_score, f1_score

# Path to your dataset
images_folder = "/kaggle/input/stanford-cars-dataset/cars_train/cars_train/"

# Load annotations
annotations_path = "/kaggle/input/cars-devkit/cars_train_annos.mat"
cars_train_annos = scipy.io.loadmat(annotations_path)

# Extract 'annotations' array
annotations_list = cars_train_annos['annotations'][0]

# Load metadata file (contains class names)
meta_path = "/kaggle/input/cars-devkit/cars_meta.mat"
cars_meta = scipy.io.loadmat(meta_path)

# Extract class names properly
class_names = [name[0] for name in cars_meta['class_names'][0]]

# Define the transformation for image preprocessing
transform = T.Compose([
    T.RandomHorizontalFlip(),
    T.RandomRotation(10),
    T.Resize((224, 224)),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Dataset class to load images and annotations
class CarsDataset(Dataset):
    def __init__(self, annotations, transform=None):
        self.annotations = annotations
        self.transform = transform

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, idx):
        anno = self.annotations[idx]
        fname = anno['fname'][0]  # Extract the filename properly as a string
        img_path = os.path.join(images_folder, fname)
        image = Image.open(img_path)

        if image.mode != 'RGB':
            image = image.convert('RGB')

        if self.transform:
            image = self.transform(image)

        label = anno['class'][0][0] - 1  # Adjusted label for correct indexing
        return image, label

# Prepare DataLoader for the dataset
dataset = CarsDataset(annotations_list, transform=transform)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# Load pretrained ViT model
model = models.vit_b_16(pretrained=True)

# Modify classifier for 196 classes (Stanford Cars dataset)
model.heads.head = torch.nn.Linear(model.heads.head.in_features, 196)

# Freeze all layers except the classifier
for param in model.parameters():
    param.requires_grad = False
for param in model.heads.head.parameters():
    param.requires_grad = True

# Set model to evaluation mode
model.eval()

# Set device for model (GPU if available, else CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Define loss function and optimizer
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)

# Training loop
num_epochs = 30
for epoch in range(num_epochs):
    running_loss = 0.0
    model.train()
    for images, labels in dataloader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    scheduler.step()

    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {running_loss/len(dataloader)}')

# Evaluate the model
all_labels = []
all_preds = []

model.eval()  # Set model to evaluation mode
with torch.no_grad():
    for images, labels in dataloader:
        images = images.to(device)
        labels = labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        all_labels.extend(labels.cpu().numpy())
        all_preds.extend(predicted.cpu().numpy())

# Calculate precision, recall, and F1 score
precision = precision_score(all_labels, all_preds, average='weighted')
recall = recall_score(all_labels, all_preds, average='weighted')
f1 = f1_score(all_labels, all_preds, average='weighted')

print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
```

## Fine-tuned

```python
import os
import torch
import numpy as np
from torch.utils.data import DataLoader, Dataset
import torchvision
import torchvision.transforms as T
from PIL import Image
import scipy.io
from torchvision import models
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import precision_score, recall_score, f1_score

# Path to your dataset
images_folder = "/kaggle/input/stanford-cars-dataset/cars_train/cars_train/"

# Load annotations
annotations_path = "/kaggle/input/cars-devkit/cars_train_annos.mat"
cars_train_annos = scipy.io.loadmat(annotations_path)

# Extract 'annotations' array
annotations_list = cars_train_annos['annotations'][0]

# Load metadata file (contains class names)
meta_path = "/kaggle/input/cars-devkit/cars_meta.mat"
cars_meta = scipy.io.loadmat(meta_path)

# Extract class names properly
class_names = [name[0] for name in cars_meta['class_names'][0]]

# Define the transformation for image preprocessing (with data augmentation)
transform = T.Compose([
    T.RandomHorizontalFlip(),
    T.RandomRotation(15),
    T.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2),  # Add color variations
    T.Resize((224, 224)),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Dataset class to load images and annotations
class CarsDataset(Dataset):
    def __init__(self, annotations, transform=None):
        self.annotations = annotations
        self.transform = transform

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, idx):
        anno = self.annotations[idx]
        fname = anno['fname'][0]  # Extract the filename properly as a string
        img_path = os.path.join(images_folder, fname)
        image = Image.open(img_path)

        if image.mode != 'RGB':
            image = image.convert('RGB')

        if self.transform:
            image = self.transform(image)

        label = anno['class'][0][0] - 1  # Adjusted label for correct indexing
        return image, label

# Prepare DataLoader for the dataset
dataset = CarsDataset(annotations_list, transform=transform)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# Load pretrained ViT model
model = models.vit_b_16(pretrained=True)

# Modify classifier for 196 classes (Stanford Cars dataset)
model.heads.head = torch.nn.Linear(model.heads.head.in_features, 196)

# Unfreeze all layers for fine-tuning
for param in model.parameters():
    param.requires_grad = True

# Set device for model (GPU if available, else CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Define loss function and optimizer
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.00005)  # Fine-tune with a lower learning rate
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)  # Learning rate decay

# Load the previous model checkpoint to continue training
checkpoint_path = 'model_checkpoint.pth'  # Path to the saved checkpoint
if os.path.exists(checkpoint_path):
    model.load_state_dict(torch.load(checkpoint_path))  # Load model weights

# Training loop
num_epochs = 50  # Continue for 50 epochs to improve further
for epoch in range(num_epochs):
    running_loss = 0.0
    model.train()
    for images, labels in dataloader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    scheduler.step()

    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {running_loss/len(dataloader)}')

    # Save the model checkpoint after each epoch
    torch.save(model.state_dict(), checkpoint_path)  # Save model after every epoch

# Evaluate the model after fine-tuning
all_labels = []
all_preds = []
correct = 0
total = 0

model.eval()  # Set model to evaluation mode
with torch.no_grad():
    for images, labels in dataloader:
        images = images.to(device)
        labels = labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        all_labels.extend(labels.cpu().numpy())
        all_preds.extend(predicted.cpu().numpy())

        total += labels.size(0)
        correct += (predicted == labels).sum().item()

# Calculate accuracy
accuracy = 100 * correct / total
print(f'Accuracy: {accuracy:.4f}%')

# Calculate precision, recall, and F1 score
precision = precision_score(all_labels, all_preds, average='weighted')
recall = recall_score(all_labels, all_preds, average='weighted')
f1 = f1_score(all_labels, all_preds, average='weighted')

print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
```
