# Train SRGNN

```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.models as models
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import scipy.io
import scipy.io as sio
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score
```

```python
# -------------------------------
# Custom Dataset for Car Images Using .mat Annotations
# -------------------------------
class CarDataset(Dataset):
    def __init__(self, root_dir, ann_file, transform=None, has_labels=True):
        """
        Args:
            root_dir (str): Directory with all the images.
            ann_file (str): Path to the .mat file containing annotations.
            transform (callable, optional): Transformations to be applied on an image.
            has_labels (bool): Whether the annotation file contains label information.
        """
        self.root_dir = root_dir
        self.transform = transform
        self.has_labels = has_labels

        # Load the .mat file
        mat = scipy.io.loadmat(ann_file)
        # Assume the annotations are stored under the key 'annotations'
        annotations = mat['annotations']
        self.samples = []

        for ann in annotations[0]:
            fname = ann['fname'][0]  # Extract filename string
            if self.has_labels:
                fields = ann.dtype.names
                if 'class' in fields:
                    label = int(ann['class'][0])
                elif 'class_id' in fields:
                    label = int(ann['class_id'][0])
                else:
                    raise ValueError("Annotation does not contain a 'class' or 'class_id' field.")
                self.samples.append((fname, label))
            else:
                self.samples.append((fname, None))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        fname, label = self.samples[idx]
        img_path = os.path.join(self.root_dir, fname)
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        if self.has_labels:
            return image, label
        else:
            return image

# -------------------------------
# Define the SRGNNModule
# -------------------------------
class SRGNNModule(nn.Module):
    def __init__(self, in_channels, hidden_dim):
        """
        Args:
            in_channels (int): Number of channels in the input feature map.
            hidden_dim (int): Dimension for the intermediate (query, key, value) representations.
        """
        super(SRGNNModule, self).__init__()
        self.fc_query = nn.Linear(in_channels, hidden_dim)
        self.fc_key   = nn.Linear(in_channels, hidden_dim)
        self.fc_value = nn.Linear(in_channels, hidden_dim)
        self.fc_out   = nn.Linear(hidden_dim, in_channels)

    def forward(self, x):
        # x: (B, C, H, W)
        B, C, H, W = x.size()
        # Flatten spatial dimensions (each spatial location becomes a node)
        x_flat = x.view(B, C, -1).permute(0, 2, 1)  # shape: (B, N, C) with N = H*W
        Q = self.fc_query(x_flat)
        K = self.fc_key(x_flat)
        V = self.fc_value(x_flat)
        # Compute attention weights: (B, N, N)
        attn = torch.bmm(Q, K.transpose(1, 2)) / (K.size(-1) ** 0.5)
        attn = F.softmax(attn, dim=-1)
        # Aggregate the values
        out = torch.bmm(attn, V)
        out = self.fc_out(out)
        # Reshape back to (B, C, H, W)
        out = out.permute(0, 2, 1).view(B, C, H, W)
        return out

# -------------------------------
# Define a Simple Model Using the SRGNNModule
# -------------------------------
class SimpleSRGNNModel(nn.Module):
    def __init__(self, num_classes, weights=True):
        """
        Args:
            num_classes (int): Number of output classes.
            weights (bool): Whether to use pretrained weights for the backbone.
        """
        super(SimpleSRGNNModel, self).__init__()
        # Use ResNet50 as a backbone (remove the final average pool and FC layers)
        backbone = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
        self.backbone = nn.Sequential(*list(backbone.children())[:-2])  # Output: (B, 2048, H, W)
        # Apply the SRGNN module on top of the backbone
        self.srgnn = SRGNNModule(in_channels=2048, hidden_dim=512)
        # Global average pooling and classification
        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))
        self.classifier = nn.Linear(2048, num_classes)

    def forward(self, x):
        x = self.backbone(x)
        x = self.srgnn(x)
        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        logits = self.classifier(x)
        return logits
```

```python
# -------------------------------
# Image Transforms
# -------------------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# -------------------------------
# Paths for Training and Testing Data & Annotations
# -------------------------------
train_root = "/content/cars_train/cars_train"
train_ann_file = "/content/drive/MyDrive/Colab Notebooks/car/car_devkit/devkit/cars_train_annos.mat"
test_root = "/content/cars_test/cars_test"
test_ann_file = "/content/drive/MyDrive/Colab Notebooks/car/car_devkit/devkit/cars_test_annos.mat"

# -------------------------------
# Create Datasets and DataLoaders
# -------------------------------
train_dataset = CarDataset(
    root_dir=train_root,
    ann_file=train_ann_file,
    transform=transform,
    has_labels=True
)
# For testing, if ground-truth labels are not available, set has_labels=False.
test_dataset = CarDataset(
    root_dir=test_root,
    ann_file=test_ann_file,
    transform=transform,
    has_labels=False
)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False, num_workers=2)

# Determine the number of classes based on training annotations (assuming labels are 1-indexed)
num_classes = max(label for _, label in train_dataset.samples)
print("Number of classes:", num_classes)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# -------------------------------
# Model, Loss Function, and Optimizer Setup
# -------------------------------
model = SimpleSRGNNModel(num_classes=num_classes, weights=False).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
# -------------------------------
# Training Loop
# -------------------------------
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images = images.to(device)
        # Convert labels from 1-indexed to 0-indexed for PyTorch
        labels = (labels - 1).to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * images.size(0)
    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}")
```

```python
# -------------------------------
# Evaluate on the Training Set Using the Official MATLAB Devkit Script
# -------------------------------
# Create an evaluation DataLoader for the training set (without shuffling to preserve order)
train_eval_loader = DataLoader(train_dataset, batch_size=8, shuffle=False, num_workers=2)
all_train_preds = []
model.eval()
with torch.no_grad():
    for images, _ in train_eval_loader:
        images = images.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        # Collect predictions (currently 0-indexed)
        all_train_preds.extend(predicted.cpu().numpy())

# Convert predictions to 1-indexed as expected by the MATLAB evaluation script
all_train_preds = np.array(all_train_preds) + 1
# Save the predictions to a .mat file
sio.savemat("train_predictions.mat", {"predictions": all_train_preds})
print("Training predictions saved to train_predictions.mat")

# Run the MATLAB evaluation script (requires MATLAB to be installed and accessible)
os.system("matlab -nodisplay -nosplash -r \"run('/content/car_devkit/devkit/eval_train.m'); exit;\"")

# -------------------------------
# Testing Loop (Without Ground-Truth Labels)
# -------------------------------
model.eval()
all_test_preds = []
with torch.no_grad():
    for images in test_loader:
        # For test dataset without labels, DataLoader returns only images.
        if isinstance(images, (list, tuple)):
            images = images[0]
        images = images.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        all_test_preds.extend(predicted.cpu().numpy())

print("Test predictions computed. No ground truth labels available for evaluation.")
np.save("test_predictions.npy", np.array(all_test_preds))
```

```python
def plot_metrics(y_true, y_pred, num_classes):
    """
    Computes accuracy, F1 score, and confusion matrix from ground truth and predictions,
    and plots them in a single figure.

    Args:
        y_true (list or np.array): True labels.
        y_pred (list or np.array): Predicted labels.
        num_classes (int): Number of classes.
    """
    # Compute metrics
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='weighted')
    cm = confusion_matrix(y_true, y_pred)

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Plot the Confusion Matrix
    ax_cm = axes[0]
    im = ax_cm.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    ax_cm.figure.colorbar(im, ax=ax_cm)
    tick_marks = np.arange(num_classes)
    ax_cm.set_xticks(tick_marks)
    ax_cm.set_yticks(tick_marks)
    ax_cm.set_xlabel("Predicted Label")
    ax_cm.set_ylabel("True Label")
    ax_cm.set_title("Confusion Matrix")

    thresh = cm.max() / 2.0
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax_cm.text(j, i, format(cm[i, j], 'd'),
                       ha="center", va="center",
                       color="white" if cm[i, j] > thresh else "black")

    # Display Accuracy and F1 Score
    ax_text = axes[1]
    ax_text.axis('off')
    metrics_text = f"Accuracy: {acc*100:.2f}%\nF1 Score: {f1:.4f}"
    ax_text.text(0.5, 0.5, metrics_text, ha='center', va='center', fontsize=16)
    ax_text.set_title("Evaluation Metrics")

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

def plot_multiclass_roc(y_true, y_proba, num_classes):
    """
    Plot the micro- and macro-averaged ROC curves for a multi-class problem.
    y_true: shape (N,) with integer labels in [0..(num_classes-1)]
    y_proba: shape (N, num_classes) with predicted probabilities
    """
    # Binarize the labels (e.g., for 196 classes => shape (N, 196))
    Y_true = label_binarize(y_true, classes=np.arange(num_classes))

    # Compute ROC curve and ROC area for each class
    fpr = {}
    tpr = {}
    roc_auc = {}

    for i in range(num_classes):
        fpr[i], tpr[i], _ = roc_curve(Y_true[:, i], y_proba[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Micro-average (aggregate the contributions of all classes)
    fpr["micro"], tpr["micro"], _ = roc_curve(Y_true.ravel(), y_proba.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    # Macro-average: average the AUCs of each class
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(num_classes)]))
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(num_classes):
        mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
    mean_tpr /= num_classes

    fpr["macro"] = all_fpr
    tpr["macro"] = mean_tpr
    roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

    # Plot micro-average ROC curve
    plt.figure()
    plt.plot(fpr["micro"], tpr["micro"],
             label=f'micro-average ROC curve (area = {roc_auc["micro"]:.4f})',
             color='deeppink', linestyle=':', linewidth=4)

    # Plot macro-average ROC curve
    plt.plot(fpr["macro"], tpr["macro"],
             label=f'macro-average ROC curve (area = {roc_auc["macro"]:.4f})',
             color='navy', linestyle=':', linewidth=4)

    # Plot ROC for each class (optional, can be very busy if num_classes is large)
    # for i in range(num_classes):
    #     plt.plot(fpr[i], tpr[i], lw=1,
    #              label=f'Class {i} (area = {roc_auc[i]:.2f})')

    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Multi-class ROC')
    plt.legend(loc="lower right")
    plt.show()

# Example usage:
# y_true: shape (N,) with integer labels
# y_proba: shape (N, num_classes) with predicted probabilities
# num_classes = 196

plot_multiclass_roc(ground_truth, predictions, num_classes=196)
```
